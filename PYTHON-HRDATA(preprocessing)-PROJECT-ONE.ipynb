```python
import pandas as pd
import numpy as np 
import matplotlib as plt



HrTrainingData = pd.read_csv ('HR Training Data.csv')
df =  HrTrainingData
df
```

```python
HrTrainingData_no_id = pd.read_csv("HRData (13).csv")
df13 = HrTrainingData_no_id
df13
```

```python
df13['EmployeeNumber'] = df13['EmployeeNumber'].astype(str)
df13['EmployeeNumber'] = df13['EmployeeNumber'].apply(lambda x: f'{int(x):09d}' if x.isdigit() else x)
df13['EmployeeNumber'] = df13['EmployeeNumber'].apply(lambda x: f'{x[:2]}-{x[2:]}' if len(x) == 9 else x)
print(df13[['EmployeeNumber']].head())
df13.head()
```

```python
HRDATA1 = pd.read_csv ("HRData (1).csv")
df1 = HRDATA1
df1
```

```python

HRDATA2 = pd.read_csv ("HRData (2).csv")
df2 = HRDATA2

HRDATA3= pd.read_csv ("HRData (3).csv")
df3 = HRDATA3

HRDATA4 = pd.read_csv ("HRData (4).csv")
df4 = HRDATA4

HRDATA5 = pd.read_csv ("HRData (5).csv")
df5 = HRDATA5

HRDATA6 = pd.read_csv ("HRData (6).csv")
df6 = HRDATA6

HRDATA7 = pd.read_csv ("HRData (7).csv")
df7 = HRDATA7

HRDATA8 = pd.read_csv ("HRData (8).csv")
df8 = HRDATA8

HRDATA9 = pd.read_csv ("HRData (9).csv")
df9 = HRDATA9

HRDATA10 = pd.read_csv ("HRData (10).csv")
df10 = HRDATA10

HRDATA11 = pd.read_csv ("HRData (11).csv")
df11 = HRDATA11

HRDATA12 = pd.read_csv ("HRData (12).csv")
df12 = HRDATA12

HRDATA14 = pd.read_csv ("HRData (14).csv")
df14 = HRDATA14
```

```python
HrData_1_through_14 = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df14]
for i, HrData_1_through_14 in enumerate(HrData_1_through_14, start=1):
    result = df.merge(HrData_1_through_14, how='inner', on='EmployeeNumber')
    if result.empty:
        print(f"No matching EmployeeNumbers between df and df{i}.")
    else:
        print(f"Matching EmployeeNumbers found between df and df{i}. Number of matches: {len(result)}")
    
```

```python
MERGED_HR_TRAINING_DATA1 = pd.concat([df,df1, df2, df3, df4, df5,
                                    df6,df7,df8,df9,df10,df11,
                                    df12,df13,df14])
df15 = MERGED_HR_TRAINING_DATA1
df15.head()
```
